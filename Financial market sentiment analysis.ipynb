**FINANCIAL MARKET SENTIMENT ANALYSIS**

Market sentimnet is the current attitude of investors overall regarding a company sector,or the financial market as a whole. the mood of the market is affected by crowd psychology. it is revealed through buying and selling activity.

Indicators of market sentimnet are VIX(volatility index),high-low index,the bullish percent index(BPI) and moving averages.

Though a handy tool in financial markets,market sentimnet has limitations,the herd mentality is swayed most easily by fear and greed.morever short-terms,events,worries and even rumors can sway market sentimnet especially in fast-paced,high-liquidity markets.

```python
news=[]
for row in range(0,len(df.index)):
  news.append(' '.join(str(x) for x in df.iloc[row,2:27]))
```

```python
X=news
```

```python
type(X)
```

**GET FEATURE TO TEXT CONVERSION TO BAG OF WORDS**

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(lowercase=True,ngram_range=(1,1))
```

```python
X= cv.fit_transform(X)
```

```python
X.shape
```

```python
y=df['Label']
```

```python
y.shape
```

**GET TRAIN TEST SPLIT**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,stratify=y,random_state=2529)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=200)
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
